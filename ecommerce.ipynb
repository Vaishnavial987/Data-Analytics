1.plot the gender wise no_of_customers.
OBSERVATION:it shows that more sales were made by female customers compared to male and others customers.

```python
plt.figure(figsize=(5,3))
sns.countplot(df,x='Gender')
plt.show()
```

2.plot the proportion of male and female and other gender.
OBSERVATION:it shows that female customers form the largest portion of total sales, followed by males

```python
g_cnt=df['Gender'].value_counts()
g_cnt.values
g_cnt.index
```

```python
plt.figure(figsize=(4,2))
plt.pie(x=g_cnt.values,colors=['green','blue',"red"],autopct='%.0f')
plt.show()
```

3.What is the Order count over a city and total sales over City?
OBSERVATION: it shows that total amnt payed by all cities and how many orders are placed in which city.

```python
fig,a=plt.subplots(nrows=2,figsize=(10,8))
sns.lineplot(df,x='City',y='Total_amnt',ax=a[0])
sns.lineplot(df,x='City',y='order_cnt',ax=a[1])
# sns.lineplot(df,x='Delivery',y='rating',ax=a[2])
plt.tight_layout()
plt.show()
```

4.compare gender-based spending patterns within each city.
OBSERVATION:it shows that in Eskisehir city more 'other' category people payed highest amnt.

```python
plt.figure(figsize=(10,6))
sns.barplot(df,x='City',y='Total_amnt',hue="Gender")
plt.show()
```

5.How does the distribution of customers by age category vary based on the number of orders placed?
OBSERVATION:It shows that young adults (age 18–35) place the highest number of orders, showing that this age group is the most engaged in online shopping.

```python
plt.figure(figsize=(8,6))
sns.histplot(df,x='age category',multiple='stack')
plt.show()
```

6.plot the heatmap  and highly dependent columns.
OBSERVATION:It shows a strong positive correlation between Total_amnt and Discount_amnt, meaning higher order amounts often get higher discounts.
There is also a moderate correlation between View_pages and Total_amnt, suggesting customers who view more pages tend to spend more


```python
cr = df.corr(numeric_only=True)
sns.heatmap(cr, annot=True, fmt='.1f')
plt.show()
```

7.Which product category has the highest median price?
OBSEVATION: It shows that Electronics have the highest price range and most outliers

```python
plt.figure(figsize=(10,6))
sns.boxplot(df,x='P_category',y='price')   
plt.show()
```

8.How does discount affect rating?
OBSERVATION:it shows that discounts do not have a strong direct impact on ratings

```python
plt.figure(figsize=(6,6))
sns.scatterplot(df,x='Discount_amnt',y='rating',hue='P_category')   
plt.show()
```

10.Does the number of pages viewed have a stronger relationship with rating or with total amount spent?
OBSERVATION:IT shows that as page views increase, total amount tends to increase 

```python
plt.figure(figsize=(6,5))
sns.pairplot(df[['Total_amnt','View_pages','rating']]) 
plt.show()
```

11.plot Gender-Based Spending Patterns.
OBSERVATION:it shows that most transactions fall in the lower to mid purchase amount range, indicating that smaller-value purchases are more common than large orders. and both graphs are right skewed

```python
f,axes=plt.subplots(1,2,figsize=(10,4))
sns.kdeplot(df,x='Total_amnt',hue='Gender',ax=axes[0])
sns.kdeplot(df,x='Discount_amnt',ax=axes[1])
plt.tight_layout()
plt.show()
```

12.Which gender shows greater diversity in device usage patterns?
OBSERVATION:It shows that Female customers use and purchase more devices compared to males

```python
fg=sns.FacetGrid(df,col='Gender')
fg.map_dataframe(sns.countplot,x='Device')
plt.show()
```
