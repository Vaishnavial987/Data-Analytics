```python
#VALUE COUNT()=it will count how many times each value is present.(return type is serie[index and column]) groping or aggregate function.
ticket_cnt=df['ticket'].value_counts()
```

```python
df['age category']=pd.cut(df['age'],bins=[0,2,15,30,59,100],labels=['infant','child','teen','adult','senior'],right=True,include_lowest=True)
df
```

```python
def comp_type(row):
    if row['fam_cnt']==1:
        return "solo"
    elif row['psg_cnt']>1 and row['fam_cnt']==0:
        return "friends"
    else: 
        return "family"
df['comp_type']=df.apply(comp_type,axis=1)#axis=1=> all columns of 1st row,2nd row,..(row by row)axis=0=>all row of 1st col then 2nd col,...(col by col)
```

9.plot the distribution of 1.age 2.fare

```python
import seaborn as sns
f,a=plt.subplots(nrows=2,figsize=(10,6))
sns.kdeplot(df,x='age',ax=a[0])
sns.kdeplot(df,x='total_fare',ax=a[1])
a[0].set_xticks(range(0,85,2))
a[0].grid()
a[1].grid()
plt.tight_layout()
plt.show()
```
