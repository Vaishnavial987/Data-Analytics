```python
import pandas as pd
```

```python
f_name='movies.csv'
df=pd.read_csv(f_name)
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
for col in df.columns:
    if df[col].isnull().any():
        cnt=df[col].isnull().sum()
        print(col,cnt,sep='-->')
```

```python
for col in['IMDb','Runtime']:
    df[col]=df[col].fillna(df[col].mean())

for col in ['Directors','Genres','Country','Language']:
    df[col]=df[col].fillna('unknown')
```

```python
df['Age'].unique
```

```python
# df['Age'].value_counts()
df['Age']=df['Age'].fillna(df['Age'].mode()[0])
```

```python
df['Rotten Tomatoes']
```

```python
df['Rotten Tomatoes']=df['Rotten Tomatoes'].str.removesuffix('%').astype('float')#missing values always in the form of float,we cant convert them as int
```

```python
df
```

```python
df['Rotten Tomatoes']=df['Rotten Tomatoes'].fillna(df['Rotten Tomatoes'].mean())
df['Rotten Tomatoes']=(df['Rotten Tomatoes']/10).round(1)
```

```python
df
```

```python
df['IMDb']=df['IMDb'].round(1)
df['Runtime']=df['Runtime'].round(1)
```

```python
df
```

```python
df['Type'].unique()
```

```python
df=df.drop(columns=['Unnamed: 0','Type','Rotten Tomatoes','ID'])
```

```python
df
```

```python
#MELT :works like unpivoting ,converts cols to rows,pivot:converts from rows to columns.
#syntax=pd.melt(dataframe,id_vars,value_vars,var_name,value_name)
pd.melt(df,id_vars=['Title'],value_vars=['Netflix','Hulu','Prime Video','Disney+'])
```
