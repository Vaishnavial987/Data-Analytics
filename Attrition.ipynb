```python
(df['hire_date'].dt.year).value_counts()
```

```python
hiring_cnt=(df['hire_date'].dt.year).value_counts().sort_index()
hiring_cnt
```

```python
term_cnt=(df.loc[df['status']=='Terminated','term_date'].dt.year).value_counts().sort_index()
term_cnt
```

```python
pd.merge(hiring_cnt,term_cnt,how='outer',left_index=True,right_index=True)
#merging with the help of columns.....on,left_on,right_on parameters
#on--->common column(same spelling)
# left_on,right_on--->df with diff col_names
# merging with the help of index--->left_index=True,right_index=True
```

```python
emp_cnt=pd.merge(hiring_cnt,term_cnt,how='outer',left_index=True,right_index=True)
emp_cnt.columns=['hirings','attritions']
emp_cnt=emp_cnt.fillna(0).astype(int)
```

```python
emp_cnt
```

```python
emp_cnt['remaining_emps_cnt']=emp_cnt['hirings']-emp_cnt['attritions']
emp_cnt['remaining_emps_cnt']
```

```python
emp_cnt['retention']=emp_cnt['remaining_emps_cnt'].cumsum()
emp_cnt['retention']
```

```python
emp_cnt
```

```python
df['status'].value_counts()
```

```python
fig,a=plt.subplots(ncols=2,figsize=(10,5))
sns.lineplot(emp_cnt,x=emp_cnt.index,y='hirings',ax=a[0],label='hirings',marker='o')
sns.lineplot(emp_cnt,x=emp_cnt.index,y='attritions',ax=a[0],label='attritions')
sns.lineplot(emp_cnt,x=emp_cnt.index,y='retention',ax=a[1])
a[0].set_label('no_of_emps')
a[0].set_label('year')
a[1].set_label('year')
plt.tight_layout
plt.show()
#
```
