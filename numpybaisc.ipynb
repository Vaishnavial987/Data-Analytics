```python
l=[300,310,250,330,320,900,1]
l
```

```python
np.percentile(l,25)
```

```python
np.percentile(l,50)
```

```python
np.percentile(l,75)
```

```python
q3=np.percentile(l,75)
q1=np.percentile(l,25)
print(q3)
print(q1)
```

```python
iqr=q3-q1
print(iqr)
```

detecting outliers

```python
lb=q1-(1.5*iqr)
ub=q3+(1.5*iqr)
print(lb,ub)
```

```python
for i in l:
    if i<lb or i>ub:
        print(i)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20,8))
sns.boxplot(x=l)
plt.xticks(range(0,900,20))
plt.show()
```
