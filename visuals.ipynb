```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'year': [2015,2015,2016,2016,2017,2017,2018,2018,2019,2019,2020,2020,2021,2021,2022],
    'product': ['A','B','A','B','A','B','A','B','A','B','A','B','A','B','B'],
    'sales': [120,480,135,460,128,490,150,520,140,530,155,550,160,580,1500],
    'marketing_spend': [12,42,15,40,13,45,18,47,14,48,17,50,19,53,90],
    'customer_rating': [3.8,4.2,2.9,3.5,4.5,3.8,2.7,4.7,4.9,3.2,3.0,4.8,2.8,4.1,2.5]
}

df = pd.DataFrame(data)

df
```

```python
plt.figure(figsize=(3,3))
sns.lineplot(df,x='year',y='sales')
```

```python
plt.figure(figsize=(6,4))
sns.lineplot(df,x='year',y='sales',label='sales')
sns.lineplot(df,x='year',y='marketing_spend',label='marketing_spend')
plt.show()
```

```python
fig,axes=plt.subplots(ncols=2,figsize=(4,4))
sns.lineplot(df,x='year',y='sales',label='sales')
sns.lineplot(df,x='year',y='marketing_spend',label='marketing_spend')
```

```python
fig,axes=plt.subplots(ncols=2,figsize=(8,3))
sns.lineplot(df,x='year',y='sales',label='sales',ax=axes[0])
sns.lineplot(df,x='year',y='marketing_spend',label='marketing_spend',ax=axes[1])
```

```python
fig,axes=plt.subplots(ncols=3,figsize=(8,3))
sns.lineplot(df,x='year',y='sales',label='sales',ax=axes[0])
sns.lineplot(df,x='year',y='marketing_spend',label='marketing_spend',ax=axes[1])
sns.lineplot(df,x='year',y='customer_rating',label='rating',ax=axes[2])
plt.tight_layout()
plt.show()
```

```python
fig,axes=plt.subplots(ncols=3,figsize=(8,3))
sns.lineplot(df,x='year',y='sales',hue='product',ax=axes[0])
sns.lineplot(df,x='year',y='marketing_spend',hue='product',ax=axes[1])
sns.lineplot(df,x='year',y='customer_rating',hue='product',ax=axes[2])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4,2))
sns.boxplot(df,x='sales')   #line in front of box called whisker(upper bound) and the line lies at the middle is called meadian and the back line of box is called lower bound. and the middle part is called asIQR(inter quartile range)
plt.show()   #width of whiskar or width of box depends on scatter,more bidth more scatter.
```

```python
plt.figure(figsize=(28,4))
sns.boxplot(df,x='sales')
plt.xticks(range(0,1000,25))
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(df,x='sales')
plt.xticks(range(0,1000,25),rotation=90)
plt.show()
```

```python
#corr()---> works on numerical_cols
#1--->100%,0.98--->98%,0.1-->10%,-0.21--->negative linear relation
cr=df.corr(numeric_only=True)
```

```python
sns.heatmap(cr)   
plt.show()
```

```python
sns.heatmap(cr,annot=True)   
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(cr,annot=True,fmt='.1f')   
plt.show()   #sales and marketing spend has stronger relationship
```
