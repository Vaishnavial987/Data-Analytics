# sort the data according to student id (asc order using sort_values())

```python
df=df.sort_values(by='id',ascending=True)
```

```python
df
```

# reset the index using reset_index()

```python
df=df.sort_values(by='id',ascending=True).reset_index(drop=True)
```

# change the gen col_name to gender using rename()

```python
df=df.rename(columns={'gen':'gender'})
```

```python
df
```

# get unique values of gender column using unique()

```python
df['gender'].unique()
```

# convert gender values to title case and repalce the incomplete values using (str.title() and replace() )

```python
df['gender']=df['gender'].str.title()
```

```python
df
```

# handle the null values of age and gender using fillna()

```python
age_avg=df['age'].mean()
```

```python
gen_mode=df['gender'].mode()[0]
```

```python
df['age']=df['age'].fillna(age_avg)
```

```python
df['gender']=df['gender'].fillna(gen_mode)
```

```python
df
```

# convert age column to intype usin astype()


```python
df=df.astype({'age':'int'})
```

```python
df
```

# remove the column location,state bcz it has more null values (use dropna() )

```python
df=df.dropna(axis=1).reset_index(drop=True)
```

# create the Result Classifications(grade) using score
(fail,first class,second class and distinction)

```python
df['score_category']=pd.cut(df['score'],bins=[1,30,40,60,85],labels=['fail','first class','second class','distinction'],include_lowest=True)
```
