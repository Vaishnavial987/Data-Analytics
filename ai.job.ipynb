```python
df['experiance'] = pd.cut(
    df['experiance'],
    bins=[0,3,7,15,40],
    labels=['Entry','Mid','Senior','Expert']
)
```

```python
q1 = df['avg_sal'].quantile(0.25)
q3 = df['avg_sal'].quantile(0.75)
iqr = q3 - q1
lower = q1 - 1.5*iqr
upper = q3 + 1.5*iqr
df = df[(df['avg_sal'] >= lower) & (df['avg_sal'] <= upper)]
```

```python
df.head()
```

```python
skill_cols = [col for col in df.columns if 'skill' in col.lower()]
```

```python
df['total_skill_score'] = df[skill_cols].sum(axis=1)

```

```python
print(skill_cols)
```

```python
df.rename(columns={
    'skill_1':'Analytical',
    'skill_2':'Technical',
    'skill_3':'AI_ML',
    'skill_4':'Digital_Tools',
    'skill_5':'Communication',
    'skill_6':'Leadership',
    'skill_7':'Creativity',
    'skill_8':'Critical_Thinking',
    'skill_9':'Adaptability',
    'skill_10':'Domain_Expertise'
}, inplace=True)
```

```python
soft = ['communication','leadership','creativity','critical_thinking','adaptability']
tech = ['analytical_skill','technical_programming','ai_ml_knowledge','digital_literacy','domain_expertise']
```

```python
soft = [col for col in df.columns if col.lower() in 
        ['communication','leadership','creativity','critical_thinking','adaptability']]

tech = [col for col in df.columns if col.lower() in 
        ['analytical_skill','technical_programming','ai_ml_knowledge','digital_literacy','domain_expertise']]
```

```python
df['soft_skill_score'] = df[soft].mean(axis=1)
df['technical_skill_score'] = df[tech].mean(axis=1)
```

```python
high_risk_jobs = df[df['risk_category'].str.lower() == 'high']

```

Conclusion:
AI exposure and automation are affecting future job safety and salaries.
Jobs with strong technical and digital skills are safer and offer better salary growth.
Repetitive and routine jobs face higher automation risk and lower long-term stability. those jobs will be replaced by ai. having both technical and soft skills is important to stay relevant in the AI-driven future.
