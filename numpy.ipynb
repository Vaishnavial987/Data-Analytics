```python
l=[56,78,98.6]
print(l)
```

```python
np.array(l)   #it gives in float beacause float is more powerfull datatype compared to int.
```

```python
l=[56,78,98.6,5+7j]    #it gives in complex datatype because complex datatype is more powerfull compared to int and float
print(l)
np.array(l)
```

```python
l=[56,78,98.6,5+7j,'python']    #it gives in their original datatype
np.array(l)
```

```python
l=[56,78,98.6,5+7j,'python','python']    #it supporting duplicate elements also
np.array(l)
```

```python
l=[56,78,98.6,5+7j,'python','python']    #list is supports indexing and slicing
np.array(l)
print(l[0])
print(l[-1])
print(l[::-1])
print(l[::2])
```

```python
arr=np.array(l)   #array also supports indexing and slicing.
print(arr)
arr[0]
```

```python
print(arr[0])     #array also supports indexing and slicing.
print(arr[-1])
print(arr[::-1])
print(arr[::2])
```
