```python
import numpy as np
import numpy.random as rnd
```

```python
rnd.randint(1,101)               #randint=random integers .it supports for only integer.
```

```python
rnd.randint(1,101,(5,))    #it gives the list of elements based on given value
```

```python
rnd.randint(1,101,(5,3))   #low,high,shape
```

```python
rnd.seed(34)            #seed not changes the random numbers.
arr1=rnd.randint(1,101,(5,3))
print(arr1)
```

```python
print(arr1[2,2])    #it gives the value based on the address.
```

```python
print(arr1[2])
print(arr1[4,2])   
print(arr1[-1,-1])
print(arr1[0,0])
print(arr1[::2])
print(arr1[::,1])
print(arr1[::,[0,2]])  #or
print(arr1[::,::2])
print(arr1[::-1,::2])  #or
print(arr1[::-1,[0,2]])

```

```python
l=['kiccha','sudeep','badsha','Abhinaya chakravarthi']
print(l)
```

```python
rnd.choice(l)
```

```python
rnd.choice(l,(3,))
```

```python
rnd.choice(l,(3,),p=[0.20,0.05,0.35,0.40])  #using probability factor
```

```python
#to avoid the duplicates we use replace=False
rnd.choice(l,(3,),replace=False)
```
