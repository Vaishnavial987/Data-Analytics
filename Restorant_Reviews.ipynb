```python
df['Hotel_Stay'] =df['Hotel_Stay'].replace(r"[\[\]\'\"]", "", regex=True)
```

```python
df['Hotel_Stay'] =df['Hotel_Stay'].replace(r"Stayed", "", regex=True)
```

VISUALS

```python
plt.figure(figsize=(6,8))
sns.histplot(df,x='Hotel_Rating')
plt.show()
```

Rating is more between 8 and 9.

```python
df['Review_Date'] = pd.to_datetime(df['Review_Date'])
```

```python
df['year']=df['Review_Date'].dt.year
```

```python
plt.figure(figsize=(4,4))
sns.countplot(df,x='year',palette='rainbow',hue='Has_Negative_Review' )
plt.show()
```

More no.of reviews was found in the year 2016 in both negative and positive Reviews.

```python
plt.figure(figsize=(4,4))
neg = df['Has_Negative_Review'].value_counts()
plt.pie(neg, labels=['No Negative', 'Has Negative'], autopct='%1.1f%%')
plt.show()
```

75% of people given no negative reviews but 25% of people given negative reviews

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df,x='Reviewer_Rating').grid()
plt.show()
```

the rating is more for 10 but it is left-skewed plot

```python
plt.figure(figsize=(7,3))
sns.boxplot(df,x='Reviewer_Rating')
plt.show()
```

the Outliers are detected from 2 to 4.4.which means very less people given below 4.4 rating.

```python
plt.figure(figsize=(4,5))
sns.scatterplot(df,x='Reviewer_Rating',y='Has_Negative_Review')
plt.show()
```

Reviewers who gave negative reviews generally provided lower ratings, while those without negative reviews gave higher ratings.

```python
fig,axis=plt.subplots(ncols=3,figsize=(15,4))
sns.lineplot(df,x='year',y='Hotel_Rating',label='Hotel ratings',ax=axis[0])
sns.lineplot(df,x='year',y='Reviewer_Rating',label='Reviewers rating',ax=axis[1])
sns.lineplot(df,x='year',y='Total_Number_of_Reviews',label='Total Reviews',ax=axis[2])
axis[0].set_xticks(range(2015,2018,1))
axis[1].set_xticks(range(2015,2018,1))
axis[2].set_xticks(range(2015,2018,1))
plt.tight_layout()
plt.show()
```

From 2015 to 2016, hotel ratings, reviewer ratings, and total number of reviews all increased, indicating improved guest satisfaction and engagement.

```python
plt.figure(figsize=(26,6))
c=sns.barplot(df,x='Hotel_Rating',y='Reviewer_Rating',errorbar=None) 
c.bar_label(c.containers[0])#for displaying number label
plt.show()
```

Mostly Higher-rated hotels receive higher individual reviewer ratings.

CONCLUSION:


Restorant with higher overall ratings consistently receive higher reviewer ratings, while negative reviews are strongly linked with lower scores.And in the year 2016 more ratings are occured.
However, in 2017, while ratings remained fairly stable, the total number of reviews decreased, suggesting a slight decline in customer participation despite consistent quality levels.
Restorant should focus on improving service quality, as the dataset shows a strong positive relationship between restorant Rating and Reviewer Rating
