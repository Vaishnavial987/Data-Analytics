**1.What is the overall distribution of users by gender?**

```python
plt.figure(figsize=(3,4))
sns.countplot(df,x='Gender')
plt.yticks(range(0,250,200))
plt.title('Gender Distribution')
plt.show()
```

**OBSERVATION**: The number of male users who active in social_media is higher than that of female and other category social media users. 

**2.Which social media platform is most popular among each gender group?**

```python
plt.figure(figsize=(9,4))
sns.countplot(df,x='platform',hue='Gender')
plt.title('social_media_usage Distribution')
plt.show()
```

**OBSERVATION**: Male users show consistently high usage across all platforms.TikTok and X (Twitter) have the highest male participation.Female users also show high activity but with some variation across platforms.The Other category has very low counts across all platforms.

**3.What is the average time spent on each platform, and how does it differ by gender?**

```python
sns.boxplot(data=df, x='platform', y='Screen_Time', hue='Gender')
plt.title("Time Spent on Social Media by Gender")
plt.show()
```

**OBSERVATION**: Across all platforms, screen time generally ranges from 3 to 8 hours, with a few outliers touching 10+ hours.Instagram and X (Twitter) are the platforms where users spend the most time

**4.what is the relationship between time spent and engagement level?**

```python
plt.figure(figsize=(8,3))
sns.scatterplot(data=df, x='Screen_Time', y='Stress_Level', hue='platform')
plt.title("Time Spent vs Stress Level")
plt.show()
```

**OBSERVATION**: A few users exceed 10 hours of screen time and also have stress level 10, representing extreme usage leading to peak stress.

**5.What is the platform wise age distribution of social media users in the dataset?**

```python
plt.figure(figsize=(9,3))
sns.kdeplot(data=df, x='Age',hue='platform')
plt.title("Age Distribution")
plt.show()
```

**OBSERVATION**: The usage profiles for all six flatforms are bimodal,they have two distinct peaks of user age: one in the younger adult range and one in the older adult range.

**6.plot the heatmap and higly dependent columns.**

```python
plt.figure(figsize=(4,4))
cr=df.corr(numeric_only=True)
sns.heatmap(cr,annot=True,fmt='.1f')   
plt.title("Correlation Heatmap")
plt.show()
```

**OBSERVATION**: The strongest corelations exist between variables related to Screen Time,Sleep Quality and Stress Level.Age has very little correlation with other variables.

**7.Which age groups prefer which social media platforms?**

```python
plt.figure(figsize=(10,3))
g = sns.FacetGrid(df, col="platform")
g.map(sns.histplot, "Stress_Level")
plt.show()
```

**OBSERVATION**: Facebook,TikTok and X(Twitter) all indicating the highest reported stress for a large number oof users on these platform.LinkdIn usage is associated with highly varied stress experiences,from low to high.YouTube shows a normal distribution.Instagram shows high peak

**8.What is the share of each age_category based on total users?**

```python
user_counts = df['age category'].value_counts()

plt.figure(figsize=(4,4))
plt.pie(user_counts, labels=user_counts.index, autopct='%1.1f%%')
plt.show()
```

**OBSERVATION**: Tik Tok is the most used platform,accounting for the largest share at 19.0%.and Instagram is the Least Used platform,representing the smallest share at 14.8%.

**9.How does screen time vary by gender across platforms?**

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x='platform', y='Screen_Time', hue='Gender', marker='o')
plt.title("Screen Time Trend by Gender Across Platforms")
plt.show()
```

**OBSERVATION**:All gender distribution have high Screen_Time in Instagram platform.Other gender category has high usage of social meadia in TikTok platform

**10.Which platform has Gender-wise highest total sleep_Quality across all users?**

```python
plt.figure(figsize=(6,4))
sns.barplot(data=df, x='platform', y='Sleep_Quality',hue='Gender')
plt.title("Total Sleep_Quality per Platform")
plt.show()
```

**OBSERVATION**:Sleep Quality is significantly affected by both the platform used and the users gender.Platforms like X(Twitter),YouTube, and Linkedln are associated with the highest reported sleep quality score.

**11.How does age distribution vary by gender?**

```python
plt.figure(figsize=(8,3))
sns.histplot(data=df, x='age category', hue='Gender', multiple='stack')
plt.title("Age Distribution by Gender")
plt.show()
```

**OBSERVATION**:The distribution of respondents across the three age categories (Elder user,Younger User and Adult) is very conistant and relatively equal in terms of total count.Female group makes up the largest portion.
