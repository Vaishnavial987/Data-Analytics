```python
a=np.array([8,9,10,11,12])
print(a)
```

```python
arr+a
```

```python
#Arthemetic Operator
print(arr+a)
print(arr-a)
print(arr/a)
print(arr//a)
print(arr%a)
print(arr*a)
```

```python
#Bitwise operator
print(arr>>a)
print(arr<<a)
print(arr&a)
print(arr|a)
print(arr^a)
```

```python
#Relational operator
print(arr>a)
print(arr<a)
print(arr==a)
print(arr!=a)
print(arr>=a)
print(arr<=a)
```

```python
# Membership Operator
print(arr in a)
print(arr not in a)
print(c in arr)
print(c  not in arr)
```
